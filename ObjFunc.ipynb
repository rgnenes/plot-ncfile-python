```python
from netCDF4 import Dataset
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
import warnings
import cartopy.crs as ccrs
```

```python
warnings.filterwarnings('ignore')
ncfile='/content/drive/MyDrive/Moselle_Basta_y.nc' #ncfile path
print('Reading nc file')
data = Dataset (ncfile)

lats=data.variables['lat'][:]
print('Extracting lat')
lons=data.variables['lon'][:]
print('Extracting lon')
time=data.variables['time'][:]
print('Extracting time')
SM_Lall=data.variables['SM_Lall'][:]
print('Extracting average soil moisture over all layers')
snowpack=data.variables['snowpack'][:]
print('Extracting snowpack')
satSTW=data.variables['satSTW'][:]
print('Extracting water level in groundwater reservoir')
recharge=data.variables['recharge'][:]
print('Extracting groundwater recharge')

#change coordinates as your wish according to your map

map=Basemap(projection='merc',
           llcrnrlon = 9.8,
           llcrnrlat = 48.8,
           urcrnrlon = 12.5,
           urcrnrlat = 51.5,
           resolution='i')  
lon, lat = np.meshgrid (lons,lats)
x,y = map(lon,lat)

c_scheme = map.pcolor(x, y, np.squeeze(SM_Lall[0,:,:]), vmin=0.7, vmax=1, cmap = 'RdBu') #1st part is printed[img. number,:,:] 
#vmin,vmax is colorbar change as your wish
#cmap = 'colour' usually 'jet' is used but change as your wish. For colours go https://matplotlib.org/stable/tutorials/colors/colormaps.html
map.drawcoastlines()
map.drawstates()
#map.drawcountries()
map.drawparallels(np.arange(-90.,120.,30.),labels=[1,0,0,0]) # draw parallels
map.drawmeridians(np.arange(0.,420.,60.),labels=[0,0,0,1]) # draw meridians

#colorbar is 2 units below

plt.title('2003 Filled') #Figure title
plt.savefig("2003_filled.jpg",dpi=330)
```
